```python
import pandas as pd
import plotly.figure_factory as ff
import plotly.express as px
import plotly.graph_objects as go
import statistics as st

df = pd.read_csv("newdata.csv")
df.head()
data = df['average'].tolist()
fig = ff.create_distplot([data],['distribution of data'])

mean = st.mean(data)
std_deviation = st.stdev(data)

fig.add_trace(go.Scatter(x = [mean,mean],y = [0,1.6],mode ='lines',name = 'mean of distribution'))
fig.show()
print(mean,'\t',std_deviation)
```

```python
import random

dataset =[]
for i in range(0,100):
  random_index = random.randint(0,len(data))
  value = data[random_index]
  dataset.append(value)
mean = st.mean(dataset)
std_deviation = st.stdev(dataset)

print(mean)
print(std_deviation)
```

```python
def random_set_of_mean(counter):
  new_list = []
  for i in range(counter):
    random_number = random.randint(0,len(data))
    value = data[random_number-1]
    new_list.append(value)
  mean = st.mean(new_list)
  return mean

def showData(mean_list):
  fig  =  ff.create_distplot([mean_list],['distribution after sampling'],show_hist = False)
  mean = st.mean(mean_list)
  fig.add_trace(go.Scatter(x = [mean,mean],y = [0,12.5],mode = 'lines',name = 'mean of distribution samples'))
  fig.show()
  print("standard deviation of sampling data is -->", st.stdev(mean_list))

def setup(data = data):
  mean_list = []
  for i in range(0,1000):
    mean = random_set_of_mean(100)
    mean_list.append(mean)
  showData(mean_list)

setup()

```
